```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from math import log2

# Define the function to calculate Shannon entropy
def shannon_entropy(probabilities):
    return -sum(p * log2(p) for p in probabilities if p > 0)

# Define the function to calculate average length using Huffman coding
def huffman_average_length(probabilities, code_lengths):
    return sum(p * l for p, l in zip(probabilities, code_lengths))

# Define probabilities for the symbols 'A' and 'B'
probabilities = [0.9, 0.1]

# Calculate Shannon entropy
entropy = shannon_entropy(probabilities)
print(f"Shannon Entropy: {entropy:.3f} bits per symbol")

# Huffman code lengths for 'A' and 'B' (based on given probabilities)
huffman_code_lengths = [1, 2]  # 'A' -> '0', 'B' -> '10'

# Calculate average length of Huffman-coded symbols
average_length_huffman = huffman_average_length(probabilities, huffman_code_lengths)
print(f"Huffman Coding Average Length: {average_length_huffman:.3f} bits per symbol")

# Plotting the results
labels = ['Shannon Entropy', 'Huffman Coding']
values = [entropy, average_length_huffman]

plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['blue', 'green'])
plt.ylabel('Bits per symbol')
plt.title('Comparison of Shannon Entropy and Huffman Coding Efficiency')
plt.ylim(0, max(values) + 0.5)
plt.show()
```

```python
import io
from misc.ref.python.arithmeticcoding import ArithmeticEncoder, BitOutputStream, SimpleFrequencyTable

# Define a function to perform arithmetic encoding
def arithmetic_encode(probabilities, symbols, sequence, numbits=32):
    # Map symbols to integer indices
    symbol_to_index = {sym: idx for idx, sym in enumerate(symbols)}
    indexed_sequence = [symbol_to_index[sym] for sym in sequence]
    
    # Create an in-memory bytes buffer
    buffer = io.BytesIO()
    bitout = BitOutputStream(buffer)
    
    # Initialize frequency table and encoder
    freqs = SimpleFrequencyTable([int(p * 1000) for p in probabilities])  # Scale probabilities to integer frequencies
    enc = ArithmeticEncoder(numbits, bitout)
    
    # Encode each symbol using their integer indices
    for idx in indexed_sequence:
        enc.write(freqs, idx)
    enc.finish()
    
    # Get the bytes from the buffer before closing the BitOutputStream
    encoded_bytes = buffer.getvalue()
    bitout.close()
    buffer.close()
    
    return encoded_bytes

# Define the symbols and their probabilities
symbols = ['A', 'B', 'C', 'D', 'E']
probabilities = [0.5, 0.2, 0.15, 0.1, 0.05]
sequence = "ABCDEABCDEABCDEABCDEABCDE"

# Perform arithmetic encoding
encoded_bytes = arithmetic_encode(probabilities, symbols, sequence)

# Calculate the average length of the encoded data
average_length_arithmetic = len(encoded_bytes) * 8 / len(sequence)
print(f"Arithmetic Coding Average Length: {average_length_arithmetic:.3f} bits per symbol")
```

```python
import numpy as np
from math import log2
import heapq

# Shannon entropy calculation
def shannon_entropy(probabilities):
    """Calculate the Shannon entropy of a probability distribution."""
    return -np.sum([p * log2(p) if p > 0 else 0 for p in probabilities])

# Huffman coding: Calculate code lengths
def huffman_code_lengths(probabilities):
    """Generate Huffman code lengths for a given probability distribution."""
    # Priority queue: Elements are tuples of (probability, count, symbol or subtree)
    queue = [(prob, i, f"Symbol{i}") for i, prob in enumerate(probabilities)]
    heapq.heapify(queue)

    count = len(probabilities)
    # Merge two least probable elements until one remains
    while len(queue) > 1:
        (prob1, _count1, left) = heapq.heappop(queue)
        (prob2, _count2, right) = heapq.heappop(queue)
        merged_prob = prob1 + prob2
        merged_tree = (merged_prob, count, (left, right))
        heapq.heappush(queue, merged_tree)
        count += 1

    # Traverse the tree to assign lengths
    def assign_lengths(tree, depth=0):
        (prob, _count, node) = tree
        if isinstance(node, str):  # It's a leaf node
            return [(node, depth)]
        else:
            left, right = node
            return assign_lengths(left, depth + 1) + assign_lengths(right, depth + 1)

    # Start with root node
    huffman_tree = queue[0]
    code_lengths = assign_lengths(huffman_tree)
    # Sort by symbol to align lengths with probabilities
    symbol_order = {f"Symbol{i}": i for i in range(len(probabilities))}
    sorted_code_lengths = sorted(code_lengths, key=lambda x: symbol_order[x[0]])
    return [length for symbol, length in sorted_code_lengths]

# Arithmetic coding: Estimate encoded data length
def arithmetic_code_length(probabilities):
    """Estimate the average length per symbol for arithmetic coding."""
    # Calculate the total encoded length using entropy
    return np.sum([p * -log2(p) for p in probabilities if p > 0])

# Define various symbol distributions
distributions = {
    "Two Symbols 90/10": [0.9, 0.1],
    "Two Symbols 70/30": [0.7, 0.3],
    "Three Symbols": [0.6, 0.3, 0.1],
    "Four Symbols": [0.5, 0.25, 0.15, 0.1]
}

# Calculate and display results for each distribution
results = {}
for name, probs in distributions.items():
    entropy = shannon_entropy(probs)
    huffman_lengths = huffman_code_lengths(probs)
    avg_length_huffman = np.sum(np.array(probs) * np.array(huffman_lengths))
    avg_length_arithmetic = arithmetic_code_length(probs)
    
    results[name] = {
        "Entropy": entropy,
        "Huffman Average Length": avg_length_huffman,
        "Arithmetic Average Length": avg_length_arithmetic
    }

# Display results
for name, result in results.items():
    print(f"{name} Distribution:")
    print(f"  Shannon Entropy: {result['Entropy']:.3f} bits per symbol")
    print(f"  Huffman Coding Average Length: {result['Huffman Average Length']:.3f} bits per symbol")
    print(f"  Arithmetic Coding Average Length: {result['Arithmetic Average Length']:.3f} bits per symbol")
```
